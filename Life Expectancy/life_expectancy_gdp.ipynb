```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
who = pd.read_csv('all_data.csv')
print(who.head())
```

```python
who.rename(columns = {'Life expectancy at birth (years)': 'LEAB'}, inplace = True)
print(who.head())
```

```python
plt.figure(figsize=(10, 8))
sns.lineplot(data= who, x = 'Year', y = 'LEAB', hue = 'Country', marker = 'o')
plt.title('Life Expectancy Over Time in Six Nations')
plt.ylabel('Life expectancy at birth (years)')
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.lineplot(data= who, x = 'Year', y = 'GDP', hue = 'Country', marker = 'o')
plt.title('Gross Domestic Product Over Time in Six Nations')
plt.ylabel('Gross Domestic Product')
plt.xticks(rotation=30)
plt.show()
```

```python
life_expectancy = who.groupby('Country')['LEAB'].mean()
print(f'The life expectancy for each of these nations is:\n{life_expectancy}')
```

```python
plt.figure(figsize=(10, 8))
sns.regplot(x = 'GDP', y = 'LEAB', data = who, scatter_kws = {'alpha':0.5})
plt.title('Correlation between GDP and Life Expectancy')
plt.xlabel('Gross Domestic Product')
plt.ylabel('Life Expectancy')
plt.show()
```

A correlation coefficient of approximately 0.343 suggests a moderate, positive linear relationship between GDP 
and Life Expectancy

```python
correlation = who['GDP'].corr(who['LEAB'])
print("Correlation coefficient:", correlation)
```
